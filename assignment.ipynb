```python
import matplotlib.pyplot as plt
import statistics as stats
```

Exercise 16

```python
#  Weighted Mean Calculation


rate_research = 100
rate_consult = 75
rate_brief = 200
hours_research = 10
hours_consult = 10
hours_brief = 20
cost_research = rate_research * hours_research
cost_consult = rate_consult * hours_consult
cost_brief = rate_brief * hours_brief

# Weighted mean formula: sum(weight * value) / sum(weight)
total_cost = cost_research + cost_consult + cost_brief
total_hours = hours_research + hours_consult + hours_brief
weighted_mean = total_cost / total_hours

print("Weighted Mean Hourly Charge = $", weighted_mean)

activities = ["Research", "Consultation", "Brief Writing"]
rates = [rate_research, rate_consult, rate_brief]
hours = [hours_research, hours_consult, hours_brief]

plt.figure(figsize=(7,5))
plt.scatter(activities, rates, s=[h*20 for h in hours], color='blue')

plt.axhline(y=weighted_mean, color='red', linestyle='--', label=f"Weighted Mean = ${weighted_mean:.2f}")

plt.title("Hourly Rates vs Activities (Size ~ Hours Worked)")
plt.xlabel("Activity")
plt.ylabel("Hourly Rate ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Data
data = [41, 15, 39, 54, 31, 15, 33]
# (a) Mean
mean_val = stats.mean(data)
# (b) Median
median_val = stats.median(data)
# (c) Mode
mode_val = stats.mode(data)

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=7, color="skyblue", edgecolor="black", alpha=0.7)
# added vertical lines for mean, median, and mode
plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, color="green", linestyle="-.", linewidth=2, label=f"Median = {median_val}")
plt.axvline(mode_val, color="purple", linestyle=":", linewidth=2, label=f"Mode = {mode_val}")

plt.title(" Changes in Last 7 Days", fontsize=14)
plt.xlabel("Number of Changes", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
