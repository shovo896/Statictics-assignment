```python
#  Weighted Mean Calculation


rate_research = 100
rate_consult = 75
rate_brief = 200
hours_research = 10
hours_consult = 10
hours_brief = 20
cost_research = rate_research * hours_research
cost_consult = rate_consult * hours_consult
cost_brief = rate_brief * hours_brief

# Weighted mean formula: sum(weight * value) / sum(weight)
total_cost = cost_research + cost_consult + cost_brief
total_hours = hours_research + hours_consult + hours_brief
weighted_mean = total_cost / total_hours

print("Weighted Mean Hourly Charge = $", weighted_mean)

activities = ["Research", "Consultation", "Brief Writing"]
rates = [rate_research, rate_consult, rate_brief]
hours = [hours_research, hours_consult, hours_brief]

plt.figure(figsize=(7,5))
plt.scatter(activities, rates, s=[h*20 for h in hours], color='blue')

plt.axhline(y=weighted_mean, color='red', linestyle='--', label=f"Weighted Mean = ${weighted_mean:.2f}")

plt.title("Hourly Rates vs Activities (Size ~ Hours Worked)")
plt.xlabel("Activity")
plt.ylabel("Hourly Rate ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
ages = [12, 8, 17, 6, 11, 14, 8, 17, 10, 8]
# Calculate
mean_age = stats.mean(ages)
median_age = stats.median(ages)
mode_age = stats.mode(ages)

print("Mean:", mean_age)
print("Median:", median_age)
print("Mode:", mode_age)
# Histogram
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(ages, bins=range(5, 19), color='skyblue', edgecolor='black', rwidth=0.8)
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_age}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_age}')
plt.title('Age Distribution - Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()

# Bar Chart (Frequency)
plt.subplot(1,2,2)
age_freq = {age: ages.count(age) for age in set(ages)}
plt.bar(age_freq.keys(), age_freq.values(), color='orange', edgecolor='black')
plt.title('Age Frequency - Bar Chart')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Analysis part 
print("\nAnalysis:")
print(f"The mean age is {mean_age}, which gives the average age of the group.")
print(f"The median age is {median_age}, which is the middle value when ages are sorted.")
print(f"The mode is {mode_age}, which indicates that {mode_age} years old is the most common age among the group.")
```

```python
initial_value = 14468000   
final_value = 95000000     
years = 2009 - 1996        

growth_rate = (final_value / initial_value) ** (1 / years) - 1
growth_rate_percent = growth_rate * 100
print(f"The geometric mean annual increase is: {growth_rate_percent:.2f}%")
years_range = np.arange(1996, 2010)
values = [initial_value * ((1 + growth_rate) ** (year - 1996)) for year in years_range]
# Visualization
plt.figure(figsize=(10,6))
plt.plot(years_range, values, marker='o', label='Estimated Growth')
plt.title('Taxpayer Growth from 1996 to 2009')
plt.xlabel('Year')
plt.ylabel('Number of Taxpayers')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# data analysis 
absolute_increase = final_value - initial_value
average_annual_increase = absolute_increase / years

print("\nAnalytical Summary:")
print(f"Initial Value (1996): {initial_value:,}")
print(f"Final Value (2009): {final_value:,}")
print(f"Absolute Increase: {absolute_increase:,}")
print(f"Average Annual Increase (Linear): {average_annual_increase:,.0f}")
```

```python
initial_value = 172.2  # CPI in 2000
final_value = 214.5    # CPI in 2009
start_year = 2000
end_year = 2009
years = end_year - start_year
growth_rate = (final_value / initial_value) ** (1 / years) - 1
growth_rate_percent = growth_rate * 100

# Future prediction for next 10 years
future_years = 10
predicted_value = final_value * ((1 + growth_rate) ** future_years)

# data prep  for plotting
years_range = np.arange(start_year, end_year + future_years + 1)
cpi_values = [initial_value * ((1 + growth_rate) ** (year - start_year)) for year in years_range]

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(years_range, cpi_values, marker='o', linestyle='-', label="CPI Projection")
plt.axvline(x=end_year, color='red', linestyle='--', label="2009 (End of given data)")
plt.title("Consumer Price Index Growth and Projection", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("CPI", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Output the results
print(f"Geometric Mean Annual Increase: {growth_rate_percent:.2f}%")
print(f"Predicted CPI after 10 years (in 2019): {predicted_value:.2f}")
```

```python
# Revenue data
years = np.arange(2002, 2010)
revenue = np.array([635.2,  # 2002
                    735.1,  # hypothetical 2003
                    850.0,  # hypothetical 2004
                    970.0,  # hypothetical 2005
                    1200.0, # hypothetical 2006
                    1600.0, # hypothetical 2007
                    2400.0, # hypothetical 2008
                    3290.0])# 2009

initial_value = revenue[0]
final_value = revenue[-1]
years_diff = years[-1] - years[0]

geometric_mean_growth = (final_value / initial_value) ** (1 / years_diff) - 1
print(f"Geometric mean annual increase (CAGR): {geometric_mean_growth*100:.2f}%")

# Heatmap
growth_rate_yearly = revenue[1:] / revenue[:-1] - 1
growth_rate_matrix = np.reshape(growth_rate_yearly, (1, len(growth_rate_yearly)))

plt.figure(figsize=(10,2))
sns.heatmap(growth_rate_matrix, annot=True, cmap="YlGnBu", cbar=True, fmt=".2f",
            xticklabels=years[1:], yticklabels=["Growth"])
plt.title("Yearly Revenue Growth Rate Heatmap")
plt.show()
```
