```python
import matplotlib.pyplot as plt
```

```python
#  Weighted Mean Calculation


rate_research = 100
rate_consult = 75
rate_brief = 200


hours_research = 10
hours_consult = 10
hours_brief = 20


cost_research = rate_research * hours_research
cost_consult = rate_consult * hours_consult
cost_brief = rate_brief * hours_brief

# Weighted mean formula: sum(weight * value) / sum(weight)
total_cost = cost_research + cost_consult + cost_brief
total_hours = hours_research + hours_consult + hours_brief
weighted_mean = total_cost / total_hours

print("Weighted Mean Hourly Charge = $", weighted_mean)



activities = ["Research", "Consultation", "Brief Writing"]
rates = [rate_research, rate_consult, rate_brief]
hours = [hours_research, hours_consult, hours_brief]

plt.figure(figsize=(7,5))
plt.scatter(activities, rates, s=[h*20 for h in hours], color='blue')


plt.axhline(y=weighted_mean, color='red', linestyle='--', label=f"Weighted Mean = ${weighted_mean:.2f}")

plt.title("Hourly Rates vs Activities (Size ~ Hours Worked)")
plt.xlabel("Activity")
plt.ylabel("Hourly Rate ($)")
plt.legend()
plt.grid(True)
plt.show()
```
